## Create another column 'Num_Lines' that is a count of how many lines pass through a station. Then sort your dataframe by this column in descending order.

```python
def num_lines(line):
    count=0
    for c in line:
        count +=1
    return count
num_lines(df.linename)
df['Num_Lines'] = df.linename.map(num_lines)
df.iloc[500:520]

#or df['Num_Lines'] = df.linename.map(lambda x: len(x))
```

## Write a function to clean a column name.

```python
def clean(col_name):
    cleaned = col_name.strip()
    return cleaned
```

```python
#This is a list comprehension. It applies your clean function to every item in the list.
#We then reassign that to df.columns
#You shouldn't have to change anything here.
#Your function above should work appropriately here.
df.columns = [clean(col) for col in df.columns] 
```

```python
#Checking the output, we can see the results.
df.columns
```

## Compare subway traffic by day of the week. Display this as a graph.

```python
df.date = pd.to_datetime(df.date)
df.head()
```

```python
df['day_of_week'] = df.date.dt.day_name()
df.head(10)
```

```python
df.columns
```

```python
df_for_graph = df.groupby(['day_of_week']).sum()
df_for_graph = df_for_graph.drop('exits', axis=1)
df_for_graph = df_for_graph.drop('Num_Lines', axis=1)
df_for_graph = df_for_graph.sort_values('entries', ascending=False)
df_for_graph

```

```python
x = df_for_graph.index
y = df_for_graph['entries']                 

plt.figure(figsize=(12,6))
plt.bar(x, y, color = "skyblue", lw=.5, ec="black")

plt.xlabel('Day of Week')
plt.ylabel('Number of passangers')

plt.title('Subway Traffic by Day of Week')
plt.show()
```

## Is there more subway traffic on a weekend or a weekday?    Be specific in comparing magnitudes.

```python
df_for_graph.reset_index(inplace=True, drop=False)
df_for_graph.columns
```

```python
df_for_graph['weekend'] = df_for_graph.day_of_week.map(lambda x: x.startswith('S'))
weekend = df_for_graph.groupby('weekend').mean()
weekend.plot(kind='bar')
```
