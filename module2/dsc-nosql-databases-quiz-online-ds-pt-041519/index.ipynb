## Question 1

What is the difference between a NoSQL and a SQL database?  Which is better?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

A SQL database is also known as a relational database. It is made of a collection of tables and rows that are related/linked to each other through certain columns. It is a very common database structure, but it demands that you have a set structure, also known as schema, for your data. It uses Structured Query Language and that makes it easy to do queries. SQL databases are scaled vertically, by increasing hardware horse-power. SQL databases are more used on transaction heavy environments, such as for example on a retail store.

On the other hand, a NoSQL database can have flexible structure, which can be document-oriented, graph, key-value, and do not need a schema to be set. It is also a faster type of database, since the data is distributed among different servers. One scales it horizontally, by increasing the number of database servers. NoSQL databases are increasingly more used for big data and real-time web applications.

## Question 3

How does MongoDB work? How is it different from a relational database?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

MongoDB is a database type that stores data in unique JSON documents, where each document can have its own structure. For example, if you are using MongoDB to store chats between customers and customer service representatives, one interaction might be short, and another one might be long and include images and files. This is different from a relational database in that it does not need a previously set schema to store the data and it supports a more rich type of data structure, whereas a RDBMS (relational database management system) is structured in traditional tables with columns and rows.






## Question 4

What is fault tolerance, and what does it have to do with Resilient Distributed Datasets?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________


Fault tolerance is the assurance that a database that is stored in several servers will be entirely avaiable even in the case that some of these servers is down at the moment of a query. Resilient Distributed Datasets (RDDs) like Hadoop and Spark deal with this isue by chopping the data into different labeled blocks, and then storing multiple different backups of each block across all the different server clusters.





## Question 5

What is MapReduce? How is it related to Hadoop?

RDDs like Hadoop or Spark make use of MapReduce. Map is the act of asking each server to run the function or query at the same time, and then Reduce is the combining of results. This method yelds much faster results as compared to running a query on a single SQL server.
